```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('mall_customers.csv')

# Check for missing values
print(data.isnull().sum())

# Visualize distributions of 'Annual Income' and 'Spending Score'
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
sns.boxplot(y='Annual Income (k$)', data=data)
plt.title('Annual Income Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(y='Spending Score (1-100)', data=data)
plt.title('Spending Score Distribution')
plt.show()

# Scaling numerical features
scaler = StandardScaler()
data[['Annual Income (k$)', 'Spending Score (1-100)']] = scaler.fit_transform(data[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
# Visualizing Gender Distribution
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution of Customers')
plt.show()

# Visualizing Age Distribution
plt.figure(figsize=(8,6))
sns.histplot(data['Age'], kde=True, bins=20)
plt.title('Age Distribution of Customers')
plt.show()

# Pairplot to visualize relationships
sns.pairplot(data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']], diag_kind='kde')
plt.show()

# Heatmap for correlation
corr = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
